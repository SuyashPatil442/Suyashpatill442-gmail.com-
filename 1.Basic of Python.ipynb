**Exercise 1: Prime Numbers**

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
def is_prime(n):

    # Prime number is greater than 1
    if n <= 1:
        return False

    # Check divisibility from 2 up to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  # bool: True if the number is prime, False otherwise

    return True

# input the user
number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

Exercise 3: Squares of Even/Odd Numbers

Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.


```python
def print_even_squares(start, end):
    """Print the squares of all even numbers within the given range.

    Args:
        start (int): The starting number of the range.
        end (int): The ending number of the range.
    """
    for number in range(start, end + 1):
        if number % 2 == 0:  # Check if the number is even
            print(f"The square of {number} is {number**2}")

def main():
    # Define the range
    start = 100
    end = 200

    # Print squares of even numbers in the specified range
    print_even_squares(start, end)

if __name__ == "__main__":
    main()
```

```python
def print_Odd_squares(start, end):

    """Print the squares of all odd numbers within the given range.

    Args:
        start (int): The starting number of the range.
        end (int): The ending number of the range.
    """
    for number in range(start, end + 1):
        if number % 2 !=0:  # Check if the number is odd
            print(f"The square of {number} is {number**2}")

def main():
    # Define the range
    start = 100
    end = 200

    # Print squares of Odd numbers in the specified range
    print_Odd_squares(start, end)


if __name__ == "__main__":
    main()
```
